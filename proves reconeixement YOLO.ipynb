```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
# Cargar modelo YOLO preentrenado
model = YOLO('yolov8s.pt')  # Puedes usar un modelo entrenado para fútbol si está disponible

# Cargar imagen
results = model('C:/Users/eloim/OneDrive/Escritorio/3 er carrera/apc/cas kaggle/open-data-master/open-data-master/shoots_img/1.jpeg')

# Extraer detecciones
for detection in results[0].boxes:
    x1, y1, x2, y2 = detection.xyxy[0]
    print(f"Objeto en coordenadas ({x1}, {y1}) a ({x2}, {y2})")
annotated_image = results[0].plot()

plt.imshow(annotated_image)
plt.show()

llista=results[0].boxes
llista2=results[0]
print(llista,llista2)
```
